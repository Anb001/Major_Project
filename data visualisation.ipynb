```python
sns.set_style("whitegrid")

# Plot distribution of OSNR values
plt.figure(figsize=(8, 5))
sns.histplot(df["OSNR_dB"], bins=10, kde=True, color="blue")
plt.title("Distribution of OSNR Values")
plt.xlabel("OSNR (dB)")
plt.ylabel("Frequency")
plt.show()

# Plot distribution of ROF values
plt.figure(figsize=(8, 5))
sns.histplot(df["ROF"], bins=10, kde=True, color="green")
plt.title("Distribution of ROF Values")
plt.xlabel("ROF")
plt.ylabel("Frequency")
plt.show()

# Plot distribution of Skew values
plt.figure(figsize=(8, 5))
sns.histplot(df["Skew_ps"], bins=10, kde=True, color="red")
plt.title("Distribution of Skew Values")
plt.xlabel("Skew (ps)")
plt.ylabel("Frequency")
plt.show()

# Count of images per modulation format
plt.figure(figsize=(6, 4))
sns.countplot(x="Modulation", data=df, palette="pastel")
plt.title("Number of Images per Modulation Type")
plt.ylabel("Count")
plt.show()

# Pairplot to visualize relationships
sns.pairplot(df, hue="Modulation", diag_kind="kde", palette="husl")
plt.show()
```
